
1. What is hypothesis testing in statistics?**
> Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine if there is enough evidence to reject a stated assumption (hypothesis).

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**
>Null Hypothesis (H₀):** A statement that there is **no effect or difference**.
Alternative Hypothesis (H₁ or Ha):** A statement that **there is an effect or difference**.

> Example: H₀: μ = 50 vs. H₁: μ ≠ 50

3. What is the significance level in hypothesis testing, and why is it important?**
>Significance Level (α):** The probability of rejecting the null hypothesis when it's actually true.
Common values: 0.05, 0.01
It sets the **threshold for the P-value** — lower α means stricter criteria to reject H₀.

4. What does a P-value represent in hypothesis testing?**
>The **P-value** is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.


5. How do you interpret the P-value in hypothesis testing?**
> If **P ≤ α**: Reject the null hypothesis (evidence supports H₁).
 If **P > α**: Fail to reject the null hypothesis (not enough evidence against H₀).


6. What are Type I and Type II errors in hypothesis testing?**
>Type I Error (α):** Rejecting H₀ when it is true (false positive).
Type II Error (β):** Failing to reject H₀ when H₁ is true (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
>One-tailed test:** Tests for effect in **one direction only** (e.g., H₁: μ > 50).
Two-tailed test:** Tests for effect in **both directions** (e.g., H₁: μ ≠ 50).


8. What is the Z-test, and when is it used in hypothesis testing?**

>A **Z-test** is used when:

  * The sample size is large (n ≥ 30)
  * Population variance is known
     It's based on the **standard normal distribution**.


9. How do you calculate the Z-score, and what does it represent?**
>Z = (X̄ - μ) / (σ / √n)**
It represents how many standard deviations the sample mean (X̄) is from the population mean (μ).

10. What is the T-distribution, and when should it be used instead of the normal distribution?**
>Use the **T-distribution** when:
Sample size is small (n < 30)
Population standard deviation (σ) is unknown

11. What is the difference between a Z-test and a T-test?**
>Z-test:** Large sample, known σ
T-test:** Small sample, unknown σ
The T-distribution has **wider tails**, accounting for more variability.

12. What is the T-test, and how is it used in hypothesis testing?**
>The **T-test** checks if the means of two groups are significantly different.
 Types: One-sample, two-sample (independent), paired-sample

13. What is the relationship between Z-test and T-test in hypothesis testing?**
>Both test means but differ in assumptions:
 Use Z-test for known variance/large samples
 Use T-test for unknown variance/small samples

14. What is a confidence interval, and how is it used to interpret statistical results?**
>A **confidence interval** gives a range of values within which the population parameter is likely to lie, with a certain level of confidence (e.g., 95%).

15. What is the margin of error, and how does it affect the confidence interval?**
>The **margin of error** is the amount added/subtracted from the sample statistic to create the interval. A larger margin = **wider interval**, lower precision.



16. How is Bayes' Theorem used in statistics, and what is its significance?
>Bayes' Theorem updates the **probability of a hypothesis** based on new evidence. It is fundamental in **Bayesian statistics**.

17. What is the Chi-square distribution, and when is it used?**
>Used for **categorical data**, especially in **tests of independence** and **goodness of fit**. It measures how observed frequencies differ from expected frequencies.

18. What is the Chi-square goodness of fit test, and how is it applied?**
>Checks if sample data fits a population with a specific distribution.
Formula: Χ² = Σ\[(O - E)² / E], where O = observed, E = expected.

19. What is the F-distribution, and when is it used in hypothesis testing?**
>The **F-distribution** is used in comparing **two variances** or in **ANOVA**. It's **right-skewed** and depends on two degrees of freedom.

20. What is an ANOVA test, and what are its assumptions?**
>ANOVA (Analysis of Variance)** checks if the means of **3 or more groups** are significantly different.
Assumptions:

* Normality
* Equal variances
* Independence


21. What are the different types of ANOVA tests?**
> **One-way ANOVA:** One independent variable
 **Two-way ANOVA:** Two independent variables
 **Repeated Measures ANOVA:** Same subjects tested over time


22. What is the F-test, and how does it relate to hypothesis testing?**
>An **F-test** compares two variances to test if they are significantly different. It's used in **ANOVA** and **regression model comparisons**.

```python
#1 Perform a Z-test for comparing a sample mean to a known population mean
from scipy.stats import norm
import numpy as np

sample = np.array([52, 48, 50, 53, 47])
population_mean = 50
std_dev = 2
n = len(sample)

z = (np.mean(sample) - population_mean) / (std_dev / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p)
```

```python
#2. Simulate data for hypothesis testing and calculate P-value
data = np.random.normal(100, 15, 50)
sample_mean = np.mean(data)
pop_mean = 100
std_dev = 15

z = (sample_mean - pop_mean) / (std_dev / np.sqrt(len(data)))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p_value)
```

```python
# 3. One-sample Z-test
def one_sample_z_test(data, pop_mean, std_dev):
    n = len(data)
    z = (np.mean(data) - pop_mean) / (std_dev / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

data = np.random.normal(102, 5, 30)
z, p = one_sample_z_test(data, 100, 5)
print("Z-score:", z)
print("P-value:", p)
```

```python
#4. Two-tailed Z-test visualization
import matplotlib.pyplot as plt

z_score = 2.1
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.3)
plt.axvline(z_score, color='blue', linestyle='--')
plt.title("Two-tailed Z-test Decision Region")
plt.xlabel("Z")
plt.ylabel("Density")
plt.show()
```

```python
#5. Type 1 and Type 2 errors visualization
def visualize_errors():
    x = np.linspace(-4, 6, 1000)
    y_H0 = norm.pdf(x, 0, 1)
    y_H1 = norm.pdf(x, 2, 1)

    plt.plot(x, y_H0, label="H0 (mean=0)")
    plt.plot(x, y_H1, label="H1 (mean=2)")
    plt.fill_between(x, y_H0, where=(x > 1.645), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, y_H1, where=(x <= 1.645), color='blue', alpha=0.3, label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.show()

visualize_errors()
```

```python
#9. Confidence interval for sample mean
from scipy import stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * se
    return mean - margin, mean + margin

sample = np.random.normal(50, 10, 40)
print("Confidence Interval:", confidence_interval(sample))
```

```python
#10. Margin of error
def margin_of_error(data, confidence=0.95):
    se = stats.sem(data)
    moe = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * se
    return moe

sample = np.random.normal(70, 15, 50)
print("Margin of Error:", margin_of_error(sample))
```

```python
#11. Bayesian inference using Bayes’ Theorem
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B

posterior = bayes_theorem(0.01, 0.99, 0.05)
print("Posterior probability:", posterior)
```
