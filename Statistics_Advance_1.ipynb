Q1. What is a random variable in probability theory ?
   > A random variable is a variable whose values are determined by the outcomes of a random phenomenon. It can be thought of as a function that assigns numerical values to outcomes in a sample space.

Q2. What are the types of random variables?
>  There are two main types:
    1 Discrete Random Variables – Take countable values (e.g., 0, 1, 2...).
    2 Continuous Random Variables – Take any value within a range (e.g., height, weight).

Q3. What is the difference between discrete and continuous distributions?
>     Discrete distributions describe the probability of discrete outcomes.
     Continuous distributions describe probabilities over intervals using a density function, and the probability of a single exact value is 0.

Q4. What are probability distribution functions (PDF)?
>    For continuous variables, the Probability Density Function (PDF) describes the relative likelihood of a value occurring in a range.
     The area under the PDF curve within a range gives the probability of the variable falling in that range.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
>    PDF shows likelihood per value (or range).  
    CDF shows cumulative probability up to a value.
    CDF(x) = P(X ≤ x)

Q6. What is a discrete uniform distribution?
>    A distribution where all outcomes are equally likely.
    Example: Rolling a fair die (P(1) = P(2) = ... = P(6) = 1/6)

Q7. What are the key properties of a Bernoulli distribution?
>   Two outcomes: Success (1) and Failure (0)
    One trial
    Parameter: p (probability of success)

Q8. What is the binomial distribution, and how is it used in probability?
>   Models the number of successes in n independent Bernoulli trials
    Parameters: n (trials), p (probability of success)
    Used in quality control, medical testing, etc.

Q9.  What is the Poisson distribution and where is it applied?
>    Models the number of events in a fixed interval of time/space given a constant average rate (λ)
    Used for rare event counts: number of emails per hour, accidents per day

Q10. What is a continuous uniform distribution?
>    Every value in an interval [a, b] is equally likely
    PDF = 1 / (b − a)

Q11. What are the characteristics of a normal distribution?
>    Symmetrical, bell-shaped curve
    Mean = Median = Mode
    Defined by mean (μ) and standard deviation (σ)

Q12. What is the standard normal distribution, and why is it important?
>   A normal distribution with mean = 0 and SD = 1
    Used for standardizing values (Z-scores) and looking up probabilities

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
>   The CLT states that the sampling distribution of the sample mean will approach a normal distribution as the sample size increases, regardless of the population’s distribution.
    It allows us to use normal distribution-based methods in hypothesis testing and confidence intervals.

Q14. How does the Central Limit Theorem relate to the normal distribution?
 >   The CLT explains why many statistics follow a normal distribution and justifies using Z-scores and confidence intervals in real-world data analysis.

Q15. What is the application of Z statistics in hypothesis testing?
 >   Used to determine if a sample mean is significantly different from a population mean.
    Helps in rejecting or accepting a null hypothesis.

Q16. How do you calculate a Z-score, and what does it represent?
 >   Z = (X-μ) / σ
    Measures how many standard deviations a value (X) is from the mean (μ)
    Useful for comparing data from different distributions


Q17. What are point estimates and interval estimates in statistics?
>    Point Estimate: A single value estimate (e.g., sample mean) of a parameter.
    Interval Estimate: A range (e.g., confidence interval) within which the parameter likely falls.

Q18. What is the significance of confidence intervals in statistical analysis?
>    They give a range of values that likely contain the true population parameter.
    A 95% CI means we’re 95% confident that the interval includes the true value.

Q19. What is the relationship between a Z-score and a confidence interval?
>    Z-scores determine the margin of error in a confidence interval.
    Example: 95% CI uses a Z of 1.96

Q20. How are Z-scores used to compare different distributions?
>    They standardize values from different distributions, allowing fair comparison on the same scale.

Q21. What are the assumptions for applying the Central Limit Theorem?
>    Samples are independent
    Identically distributed
    Sample size is large enough (usually n ≥ 30)

Q22. What is the concept of expected value in a probability distribution?
>    It’s the average outcome if the experiment is repeated many times.

Q23. How does a probability distribution relate to the expected outcome of a random variable?
>    The expected value is computed using the probability distribution, weighting each value by its probability.
